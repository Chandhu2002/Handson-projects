```python
from sklearn.linear_model import LogisticRegression
```

```python
#Predict the churn on the basis of monthly charges

X = data[['MonthlyCharges']]
Y = data[['Churn']]

X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.35,random_state=1)  #0.65 -> 65% is for training purpose
x_train
y_train
```

```python
Logistic_reg = LogisticRegression()
Logistic_reg.fit(x_train,y_train)
#buiding our model -> training data -> x_train and y_train
```

```python
#prediction ->

model_predict = Logistic_reg.predict(x_test)
```

```python
model_predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test,model_predict)
acc
```
